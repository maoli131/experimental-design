# Capacity Progression

This uses the output from Brainome to plot the capacity progression curve of Fashion-MNIST dataset.

Brainome's output:
```
Capacity progression: 
at [ 5%, 10%, 20%, 40%, 80%, 100% ]
Ideal Machine Learner:  13,  14,  15,  15,  16,  16 Number of decision points
```

````python
import matplotlib.pyplot as plt
import seaborn as sns
x = [5, 10, 20, 40, 80, 100]
y = [13, 14, 15, 15, 16, 16]
plt.title("Capacity Progression")
sns.lineplot(x=x, y=y)
sns.scatterplot(x=x, y=y)
plt.xlabel("% Training Data used")
plt.ylabel("Number of Decision Points");
````
